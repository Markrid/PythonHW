```python
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
key = {'whoAmI': {"robot": 0, "human": 1}}
data = data.replace(key)
data
```

```python
# Вариант 2
import numpy as np
data["ROBOT"] = np.where(data["whoAmI"].str.contains("robot"), 1, 0)
data["HUMAN"] = np.where(data["whoAmI"].str.contains("human"), 1, 0)
data
```
